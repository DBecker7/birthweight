# Factors Affecting Birthweight

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

birth = pd.read_csv("https://raw.githubusercontent.com/DBecker7/DBecker7/main/teachingDataSets/Birthweight_reduced_R.csv")
```

```python
y = birth[["Birthweight"]]
x = birth[["Gestation"]]

myreg = LinearRegression().fit(x, y)
```

```python
y_preds = myreg.predict(x)

resids = y - y_preds

plt.scatter(resids, y_preds)
plt.show()
```

```python
X = birth[["smoker", "Gestation", "motherage", "headcircumference"]]

myreg = LinearRegression().fit(X, y)
mypreds = myreg.predict(X)
myres = y - mypreds

plt.scatter(myres, mypreds)
plt.show()
```
